# Correlation with heatmap

A correlation heatmap uses colored cell , typically in a monochromatic scale. to show a 2D correlation matrix(table) between two discrete dimensions or event types.It is very important in feature selection


Heatmap is used to find inter-relation between Each and Every Feature
inter-relation find out when values are int or float 
Not for categorical

# Correlation
A correlation Matrix is basically a covariance matrix. Also known as the auto-covariance matrix, dispersion matrix, variance matrix, or variance-covariance matrix. It is a matrix in which the i-j position defines the correlation between the ith and jth parameter of the given data set.

```python
df.dtypes
```

```python
df.corr()
```

```python
list = [[ 36, 75.2, 54280.20],  
        [ 38, 74.32, 34280.30],  
        [ 31, 70.56, 84280.50], 
        [ 34, 80.30, 44280.80],  
        [ 40, 100.03, 45280.30], 
        [ 33, 72.99, 70280.25],  
        [ 42, 85.84, 25280.75]] 
df = pd.DataFrame(list, columns =['Age', 'Weight', 'Salary']) 
display(df) 
```

# Join plot
A joint plot allow to study the relationship between 2 numeric variables. The central chart display their correlation. 
It is usually a scatterplot , a hexbin plot . 2d hsitogram or 2d density plot


Bascially used for univariant

# Univariate Analysis

```python
sns.jointplot(x='Age',y='Weight',data=df,kind='hex') ## hex=hexagon
```

```python
df=sns.load_dataset("tips")
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg') ## reg-->regression
```

# categorical plots
seaborn also help us in doing the analysis on categorical data points . In this section we will discuss about

#### violinplot
#### countplot

```python
#countplot
```

```python
df=sns.load_dataset("tips")
sns.countplot(data=df,x='sex')
```

```python
sns.countplot(data=df,x='day')
```

```python
## Barplot
## have to give both x and y values
sns.barplot(data=df,x='day',y='tip')
```

## Box plot
a box and whisker plot sometimes called a boxplot is a graph that presents information from a five number summary

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
## 25 50 57 100 some are outliers
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

## violinplot
Violin plot help us to see both the distribution of data in term of kernel density estimation and box plot

```python
sns.violinplot(x='sex',y='total_bill',data=df)
```
