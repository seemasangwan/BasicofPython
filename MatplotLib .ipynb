# MatplotLib 
Matplotlib is a low level graph plotting library in python that serves as a visualization utility.
matplotlib is plotting library for the python programming and its numerical mathematics extension Numpy. It provide and object oriented API for embedding plots into application using general purpose GUI toolkit like Tkinter , wxPython , Qt or GTK+

some major pros of matplotlib are:
Generally easy to get started for simple plot
Support for custom lables and texts 
Great control of every elements in a figure 
High quality output in many formats
very customizable in general

 # pyplot submodule and import plt alias

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
## simple example
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
## plotting using matplotlib
## plt scatter  use to scatter the values of x and y
plt.scatter(x,y,c='b') ## first is x 2nd is y and third is color
## if i have not write %matplotlib inline above then
## i have to write plt.show ()
plt.show()
```

```python
plt.plot(x,y,c='g')
## plot() function draw points draw line
```

```python
## plot without line
plt.plot(x,y,'o')
```

```python
x=np.array([1,2,4,6,8,10])
y=np.array([4,8,6,2,4,10])
plt.plot(x,y)
```

```python
## default x points
y=np.array([2,4,6,8,6,4,2])
plt.plot(y,c='r')
```

```python
## draw line as well as show points
plt.plot(y,c='g',marker='o')
```

```python
## marker can used as '*'
plt.plot(a,b,marker='*',c='b')
```

```python
## to get title like x-axis or y-axis use label
plt.scatter(a,b,c='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph in 2d')
plt.savefig('test.png')
```

```python
## use : for dotted line
## use _ for solid line
## use _ _ for dashed line
## use _. dashe/dotted
plt.plot(a,b,':',c='r')
```

```python
plt.plot(a,b,'b:*',markersize=10,mfc='r')
```

```python
## linestyle='dashed' short ls=':'
## linestyle='dotted' short ls='_ _'
## linestyle ='solid' short ls='_' or dashdot _.
## line color ='red' or c='r'
## linewidth='10.4'
```

# Label

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
## give title x and y 
## font of label
## position of label
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Sports Watch Data", fontdict = font1,loc='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.plot(x,y,'r:*',ms=20)
```

```python
plt.plot(x,y)
plt.grid()
## we can also specify which grid  plt.grid(axis='x')
plt.grid(axis='x',linestyle=':',linewidth='2',c='r')
```

# subplot() function
With the subplot() function you can draw multiple plots in one figure:

```python
plt.subplot(2,2,1) ## first-->row 2nd->column 3rd ->postion 1
plt.plot(x,y,'g')
plt.subplot(2,2,2) ## first-->row 2nd->column 3rd ->postion 2
plt.plot(x,y,'r:')
plt.subplot(2,2,3) ## first-->row 2nd->column 3rd ->postion 3
plt.plot(x,y,'b *',)
plt.subplot(2,2,4) ## first-->row 2nd->column 3rd ->postion 3
plt.plot(x,y,'c:*')
## give 2 row and 2 col just have 4 position
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

# Bars

```python
x = [2,8,10]
y = [11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,color='red')
plt.bar(x2,y2,color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('bar graph')
```

# Histogram

```python
## density and count or freq on y axis
a=np.array([10,11,12,13,14,20,25,30,30,30,25,20,10,29,20,11])
plt.hist(a,color='red')
plt.title('histogram')
```

# Pie chart

```python
y = np.array([35, 25, 25, 15])
labels='numpy','pandas','matplotlib','seaborn'
colors=['yellow','Green','Blue','lightcoral']

explode=(0.1,0,0,0) ## explode its first slice
plt.pie(y,labels=labels,colors=colors,explode=explode,autopct='%1.1f%%',shadow=True)
```

# boxplot

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
```
